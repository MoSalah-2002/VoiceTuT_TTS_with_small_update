# 𓋹 VoiceTut-TTS — Quickstart

Egyptian-Arabic & code-switching TTS. This notebook installs the package, loads the model, and synthesizes speech with a built-in voice.

> Runtime → **Change runtime type → GPU (T4 is fine)**.

```python
# 1. OmniVoice backbone (not on PyPI), then VoiceTut-TTS
!pip install -q git+https://github.com/k2-fsa/OmniVoice.git
!pip install -q voicetut-tts
```

```python
from voicetut_tts import VoiceTutTTS

tts = VoiceTutTTS.from_pretrained("MohammedAly22/voicetut-tts")
print("Built-in speakers:")
for s in tts.list_speakers():
    print(f"  {s.speaker_name:14} {s.gender:7} {', '.join(s.tags)}")
```

```python
from IPython.display import Audio

# Built-in speaker (pure Egyptian)
wav = tts.synthesize("ازيك عامل ايه النهاردة؟ يا رب تكون كويس وكله تمام.",
                     speaker="Mohamed", output="out.wav")
Audio("out.wav")
```

```python
# Code-switching + generation params
tts.synthesize("عندي meeting الساعة 3:30 ومعايا ال presentation بتاعتي",
               speaker="Asmaa", num_step=48, guidance_scale=2.5, speed=1.05,
               output="cs.wav")
Audio("cs.wav")
```

```python
# True streaming for long text
long_text = ("اهلا بيكم في حلقة جديدة من البودكاست. النهاردة هنتكلم عن موضوع مهم جدًا. "
             "الموضوع ده بيخص ال productivity وازاي تنظم وقتك. يلا نبدأ على طول!")
tts.synthesize_long(long_text, "long.wav", speaker="Sayed")
Audio("long.wav")
```
