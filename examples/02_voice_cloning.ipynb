## 1. Text normalization (no GPU needed)

```python
from voicetut_tts import ArabicNormalizer
norm = ArabicNormalizer()
for t in [
    "عندي 250 جنيه والميعاد 3:30 يوم 14/3/2024",
    "اتصل على 01001234567 او ابعت على ahmed.ali@gmail.com",
    "خصم 25% على المنتجات على https://shop.com النهاردة",
]:
    print("IN :", t)
    print("OUT:", norm(t)); print('-'*60)
```

## 2. Load the model

```python
from voicetut_tts import VoiceTutTTS
from IPython.display import Audio
tts = VoiceTutTTS.from_pretrained("MohammedAly22/voicetut-tts")
```

## 3. Record / upload a reference and clone it

```python
from google.colab import files
up = files.upload()                 # upload a 3-10s wav of any voice
ref_path = next(iter(up))

tts.synthesize("النهارده الجو حلو اوي، يلا نطلع نتمشى شوية",
               ref_audio=ref_path, ref_text=None,   # None => auto-transcribe
               output="clone.wav")
Audio("clone.wav")
```

## 4. Custom lexicon — fix a word's pronunciation globally

```python
tts.add_lexicon({"تيوت": "تُوت", "نايل": "نَايِل"})
tts.synthesize("فويس تيوت احسن موديل للغة المصرية", speaker="Mohamed", output="lex.wav")
Audio("lex.wav")
```
